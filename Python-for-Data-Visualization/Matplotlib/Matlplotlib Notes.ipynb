```python
import numpy as np
x = np.linspace(0,5,11)
y = x **2
print(x)
print(y)
```

### Functional method

```python
plt.plot(x, y)
plt.show() #not necessary
```

Add matlab-like arguments

```python
plt.plot(x,y,'r--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) #(nrows, ncols, nplot)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes.plot(x,y)
axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height in pct
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger plot")

axes2.plot(y,x)
axes2.set_title("Smaller plot")
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2) #tupple unpacking for (fig,axes)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2) #tupple unpacking for (fig,axes)
plt.tight_layout()

# axes object is a list
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2) #tupple unpacking for (fig,axes)
plt.tight_layout()

axes[0].plot(x,y)
axes[1].plot(y,x,'b.')
```

### Figure Size and DPI

```python
fig  = plt.figure(figsize=(3,2)) #(width,height)inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig  = plt.figure(figsize=(8,2)) #(width,height)inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(4,5))

axes[0].plot(x,y)
axes[1].plot(x,y)
plt.tight_layout()
```

### Labels, titles and legend

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title("The title")
ax.set_ylabel("y label")
ax.set_xlabel("x label")
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X squared')
ax.plot(x,x**3, label='X cubed')

ax.legend(loc='best')
#ax.legend(loc='center right')
#ax.legend(loc='center')
#ax.legend(loc=(0.1,0.8))
```

### Plot Appearance

#### Color

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='orange')
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green')
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FFFC00')
```

#### Line

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', linewidth=2)
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', linewidth=0.5)
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3, alpha=0.3)
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3, linestyle='--')
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3, linestyle='steps')
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3, ls='-.')
```

#### Markers

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3, ls='-', marker='o')
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=1, ls='-', marker='x')
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=1, ls='-', marker='*')
```

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=1, ls='-', marker='o', markersize=10, 
        markerfacecolor='orange', markeredgewidth=2, markeredgecolor='green')
```

####  Axes

```python
fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
